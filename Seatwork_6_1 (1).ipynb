```python
import pandas as pd
import statistics as stat
import matplotlib.pyplot as plt

filepath = '/content/AI On-Campus Research Survey (Responses).csv'

dataFrame = pd.read_csv(filepath)
```

```python
#Identify Columns
print(dataFrame.columns.tolist())
```

```python
#Take a sample from the data set
dataFrame[:20]
```

```python
#Check if there are NaN values in the DataFrame
checkNaN = dataFrame.isnull().values.any()
print(checkNaN)
```

```python
#Take last 20 elements
dataFrame.tail(20)
```

```python
#Take a sample from students who uses artificial intelligence for personal use
aiUse = dataFrame.iloc[:,2]
sample = aiUse[:20]
sample
```

```python
#Tally the number of ratings
sampleCounts = sample.value_counts()
print(sampleCounts)
```

```python
#Multiply two columns
sampleCountsDf = sampleCounts.reset_index()
sampleCountsDf.columns=['rating', 'choices']

sampleCountsDf['multiplier'] = sampleCountsDf['rating'] * sampleCountsDf['choices']
sampleCountsDf
```

```python
#Take the mean average of the people who uses Artificial intelligence for personal use
def findMean(sampleCountsDf):
  mean = stat.mean(sampleCountsDf['multiplier'])
  return mean

print(findMean(sampleCountsDf))
```

```python
#Take a sample of 20 from people who uses Artificial intelligence for school use
#aiUse = dataFrame.iloc[:,3]
#sample = aiUse[:20]

schoolUse = dataFrame.iloc[:,3]
sample2 = schoolUse[:20] #Take 20 sample for this column
sample2
```

```python
sampleCounts = sample2.value_counts()

sampleCounts2Df = sampleCounts.reset_index()
sampleCounts2Df.columns = ['rating', 'choices']
sampleCounts2Df['multiplier'] = sampleCounts2Df['rating'] * sampleCounts2Df['choices']
sampleCounts2Df
```

```python
#Find the mean of the second column
print(findMean(sampleCounts2Df))
```

```python
#Take a sample from students who are knowledgable in AI
knowledgable = dataFrame.iloc[:,1]
sample3 = knowledgable[:20]
sample3
```

```python
sampleCounts = sample3.value_counts()
sampleCounts3Df = sampleCounts.reset_index()
sampleCounts3Df.columns = ['rating', 'choices']
sampleCounts3Df['multiplier'] = sampleCounts3Df['rating'] * sampleCounts3Df['choices']
sampleCounts3Df
```

```python
print(findMean(sampleCounts3Df))
```

```python
#Functions for other measures of central tendency

def findMedian(sampleCounts):
  median = stat.median(sampleCounts['multiplier'])
  return median

def findMode(sampleCounts):
  mode = stat.mode(sampleCounts['multiplier'])
  return mode
```

```python
#Find Median
print(findMedian(sampleCountsDf))
print(findMedian(sampleCounts2Df))
print(findMedian(sampleCounts3Df))


#Find Mode
print(findMode(sampleCountsDf))
print(findMode(sampleCounts2Df))
print(findMode(sampleCounts3Df))
```

```python
#Functions for measures of dispersion
from scipy import stats as st
def findDev(sampleCounts):
  stdev = stat.stdev(sampleCounts['multiplier'])
  return stdev

def findVar(sampleCounts):
  variance = stat.variance(sampleCounts['multiplier'])
  return variance

def IQR(sampleCounts):
  IQR = st.iqr(sampleCounts['multiplier'])
  return IQR
```

```python
#Take the measures of dispersion

#Find standard deviation
print(findDev(sampleCountsDf))
print(findDev(sampleCounts2Df))
print(findDev(sampleCounts3Df))

#Find variance
print(findVar(sampleCountsDf))
print(findVar(sampleCounts2Df))
print(findVar(sampleCounts3Df))

#Find IQR
print(IQR(sampleCountsDf))
print(IQR(sampleCounts2Df))
print(IQR(sampleCounts3Df))
```
