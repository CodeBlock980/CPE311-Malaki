```python
import random

random.seed(0)

salaries = [round(random.random() * 1000000, -3) for _ in range(100)]

print(salaries)
```

# Calculating for Mean

```python
#Compute for mean without imports

def meanCal(salaries):
  mean = (sum(salaries)/len(salaries))
  return mean

print("The Mean is:",meanCal(salaries))

```

```python
#Compute for mean with import

import statistics as stat

def meanCal(salaries):
  mean = stat.mean(salaries)
  return mean

print("The Mean is:",meanCal(salaries))

```

# Calculating for Median

```python
#Compute for median without import

def medianCal(salaries):
  sortedTable = sorted(salaries)
  n = len(sortedTable)
  midVal = round(((n/2) + 1))

  return (sortedTable[midVal-1] + sortedTable[midVal - 2])/2

print("The Median is:",medianCal(salaries))



```

```python
#Compute for median with import

def medianCal(salaries):
  sortedTable = sorted(salaries)
  median = stat.median(sortedTable)
  return median
print("The Median is:",medianCal(salaries))



```

# Calculating for Mode

```python
#Ungrouped without statistics import

def modeCal(salaries):
  return max(set(salaries), key = salaries.count)

print(modeCal(salaries))


```

```python
#Ungrouped with statistics import

def modeCal(salaries):
  mode = stat.mode(salaries)
  return mode
print(modeCal(salaries))
```

# **Calculating for Sample Variance**

```python
#Sample Variance ungrouped without imports (Using Sample)

def findVar(salaries):
  sampleData = []
  for money in range(25):
    sampleData.append(salaries[money])

  n = len(sampleData)
  mean = (sum(sampleData)/n)

  initVar = 0

  for salary in sampleData:
    initVar += (((salary - mean)**2)/(n-1))
  return round(initVar,4)


print(findVar(salaries))


```

```python
#Sample Variance ungrouped with imports (Using Sample)
def findVar(salaries):
  sampleData = []
  for money in range(25):
    sampleData.append(salaries[money])

  SampleVar = stat.variance(sampleData)
  return round(SampleVar,4)

print(findVar(salaries))

```

# **Calculating for the Standard Deviation**

```python
def findDev(salaries):
  sampleData = []
  for money in range(25):
    sampleData.append(salaries[money])

  n = len(sampleData)
  mean = (sum(sampleData)/n)

  rootSquare = [(x - mean) ** 2 for x in sampleData]

  variance = sum(rootSquare)/(n-1)

  return round(variance ** 0.5, 4)

print(findDev(salaries))
```

```python
def findDev(salaries):
  sampleData = []
  for money in range(25):
    sampleData.append(salaries[money])
  stdDev = stat.stdev(sampleData)
  return round(stdDev, 4)

print(findDev(salaries))
```

# **Exercise 2**

```python
#Range
def findRange(salaries):
  range = max(salaries) - min(salaries)
  return range

print(findRange(salaries))
```

```python
#Coefficient of Variation

def findDev(salaries):
  stdDev = stat.stdev(salaries)
  return round(stdDev, 4)

def findCV(salaries,stdev):
  mean = (sum(salaries)/ len(salaries))
  cv = (stdev/mean)

  return cv

stdev = findDev(salaries)
print(findCV(salaries, stdev))
```

```python
#Interquartile Range (Population)
from scipy import stats

def findIQR(salaries):
  IQR = stats.iqr(salaries)
  return IQR

print(findIQR(salaries))
```
