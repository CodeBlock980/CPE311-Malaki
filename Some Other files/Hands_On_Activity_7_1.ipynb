```python
#Combine all dataFrames into 1

import pandas as pd

apl = pd.read_csv('/content/aapl.csv')
fbk = pd.read_csv('/content/fb.csv')
amzn = pd.read_csv('/content/amzn.csv')
nflx  = pd.read_csv('/content/nflx.csv')
ggle = pd.read_csv('/content/goog.csv')

```

```python
apl['ticker'] = ['APL' for i in range(len(apl.iloc[:,0]))]
apl
```

```python
fbk['ticker'] = ['FBK' for i in range(len(apl.iloc[:,0]))]
fbk
```

```python
amzn['ticker'] = ['AMZN' for i in range(len(amzn.iloc[:,0]))]
amzn
```

```python
nflx['ticker'] = ['NFLX' for i in range(len(nflx.iloc[:,0]))]
nflx
```

```python
ggle['ticker'] = ['GGLE' for i in range(len(ggle.iloc[:,0]))]
ggle
```

```python
response = rq.get('https://data.cms.gov/provider-data/api/1/datastore/query/xubh-q36u/0')
response.json().keys()
```

```python
import pandas as pd

# Take results in json format
responses = response.json()['results']

# Convert JSON to DataFrame
dataFrame = pd.DataFrame(responses)

# Write DataFrame to CSV
dataFrame.to_csv('hospitals.csv', index=False)

```

```python
dataFrame
```

```python
#Lets examine the data types of each columns
dataFrame.dtypes
```

```python
#Check if there are NaN and Null type in all columns.
def filterNaNull(dataSet):
  return dataSet.isna().any()

filterNaNull(dataFrame)
```

```python
#Since there was no empty value we check our dataFrame again to see if there are any duplicates

dataFrame.drop_duplicates(inplace = True)
dataFrame
```

```python
#Now We could start by making facility_id as our index
dataFrame
```

```python
dataFrame.columns
```

```python
#Since we are only tasked to find the name of the hospitals, its address and its contact information we can eliminate a lot of the columns we dont need from this dataset
dataFrame.drop(columns = dataFrame.iloc[:,7:39], inplace = True)
dataFrame
```

```python
#Lastly, we could make a full address by simply concatenating columns : address, citytown,state,zip_code, and countyparish
dataFrame['Full_Address'] = dataFrame['address'] + ' ' + dataFrame['citytown'] + ' ' + dataFrame['countyparish'] + ' ' + dataFrame['state'] + ' ' + dataFrame['zip_code']
dataFrame.drop(columns = dataFrame.iloc[:, 1:7])
```
